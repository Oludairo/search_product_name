```python
def strip_dots(text):
    return text.split('...')[0]

df['product_names'] = df['product_names'].apply(strip_dots)
```

```python
input_string = input("Enter the input string: ")

# Function to find exact matching queries
def find_exact_matching_queries(input_string, df):
    matches = []
    for product_name in df['product_names'].to_list():
        if product_name.replace(" ", "") == input_string.replace(" ", ""):
            matches.append(df.loc[df['product_names'] == product_name, 'search_query'].values[0])
    return matches[0]

matches = find_exact_matching_queries(input_string, df)
```
